```python
gamers_df = df.query('days_in_games > 0').copy()
gamers_df.shape
```

```python
share_of_gamers = len(gamers_df) / len(df) * 100
share_of_gamers
```

```python
ax = plt.pie([len(df), len(gamers_df)], autopct='%1.1f%%')
plt.legend(['Не играют', 'Играют'])
plt.title('Доля пользователей, играющих в игры')
plt.show()
```

```python

bins = [16, 16-8, 25, 35, 45, 55, float('inf') ]
labels = ['до 16', '16-17' '18-25', '26-35', '36-45', '46-55', '56+']

gamers_df['age_group'] = pd.cut(gamers_df['age'], bins=bins, labels=labels, right=False)
```

```python
sns.histplot(gamers_df['age_group'], stat='percent')

plt.title('Распределение возраста в процентах')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

age_group_counts = gamers_df['age_group'].value_counts()

total_gamers = age_group_counts.sum()
age_group_percentages = (age_group_counts / total_gamers) * 100

percentages_df = pd.DataFrame({'age_group': age_group_percentages.index, 'percentage': age_group_percentages.values})

sns.barplot(x='age_group', y='percentage', data=percentages_df)

plt.title('Распределение возраста в процентах')

plt.show()
```
